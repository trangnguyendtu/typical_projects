### get movie id & movie title & movie rating, skip the timestamp at the last column (no use of timestamp)

```python
ratingCols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=ratingCols, usecols=range(3), encoding="ISO-8859-1")

movieCols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=movieCols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
```

```python
ratings.head(25)
```

```python
ratings.groupby('title')['rating'].describe()
```

```python
userRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
```

```python
userRatings.head(25)
```

```python
ratingsMean = ratings.groupby('title')['rating'].describe()['mean']
```

```python
ratingsMean
```

## use corr() method of pandas, compute correlation score of every pair of movies

```python
corrCompute = userRatings.corr()
corrCompute.head()
```

### restrict the result with min_periods to keep only pairs rated by at least 100 users, ensuring more popular and recognizable results

```python
corrCompute = userRatings.corr(method='pearson', min_periods=100)
```

```python
corrCompute.head()
```

## add my own rating to the dataset to check if the recommended system is correct
### user_id = 0, love Titanic, Gone with the wind, Addicted to love, Forrest Gump, and hate Star Wars

```python
myRatings = userRatings.loc[0].dropna()
myRatings
```

### get similar movies with the correlation matix

```python
similarCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    sims = corrCompute[myRatings.index[i]].dropna()
    sims = sims.map(lambda x: x * myRatings[i])
    similarCandidates = pd.concat([similarCandidates, sims])
    
similarCandidates = similarCandidates.groupby(similarCandidates.index).sum()
similarCandidates.sort_values(inplace = True, ascending = False)
similarCandidates.drop(myRatings.index).head()
```
