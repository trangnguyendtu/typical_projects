```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO
from sklearn import preprocessing, svm

%matplotlib inline
```

```python
masses_data = pd.read_csv("mammographic_masses.data")

masses_data.head()
```

### 6 Attributes in total (1 goal field, 1 non-predictive, 4 predictive attributes)

1. BI-RADS assessment: 1 to 5 (ordinal, non-predictive!)  
2. Age: patient's age in years (integer)
3. Shape: mass shape: round=1 oval=2 lobular=3 irregular=4 (nominal)
4. Margin: mass margin: circumscribed=1 microlobulated=2 obscured=3 ill-defined=4 spiculated=5 (nominal)
5. Density: mass density high=1 iso=2 low=3 fat-containing=4 (ordinal)
6. Severity: benign=0 or malignant=1 (binominal, goal field!)

```python
masses_data = pd.read_csv(
    "mammographic_masses.data",
    na_values=["?"],
    names=["BI-RADS", "age", "shape", "margin", "density", "severity"],
)
```

```python
masses_data.head()
```

### drop missing values on the dataframe

```python
masses_data.dropna(inplace=True)
masses_data.describe()
```

```python
feature_values = masses_data[["age", "shape", "margin", "density"]].values
classe_values = masses_data["severity"].values
feature_names = ["age", "shape", "margin", "density"]
```

```python
scaler = preprocessing.StandardScaler()
feature_values_scaled = scaler.fit_transform(feature_values)
```

```python
feature_values_scaled
```

# Linear SVM

```python
from sklearn.model_selection import cross_val_score
```

```python
C = 1.0
svc = svm.SVC(kernel="linear", C=C)
cvs = cross_val_score(svc, feature_values_scaled, classe_values, cv=10)
```

```python
cvs.mean()
```

# Non-linear SVM 

```python
C = 1.0
svc = svm.SVC(kernel="rbf", C=C)
cvs = cross_val_score(svc, feature_values_scaled, classe_values, cv=10)
```

```python
cvs.mean()
```

```python
C = 1.0
svc = svm.SVC(kernel="sigmoid", C=C)
cvs = cross_val_score(svc, feature_values_scaled, classe_values, cv=10)
```

```python
cvs.mean()
```

```python
C = 1.0
svc = svm.SVC(kernel="poly", C=C)
cvs = cross_val_score(svc, feature_values_scaled, classe_values, cv=10)
```

```python
cvs.mean()
```

# Decision Tree

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
np.random.seed(42)

(training_inputs, testing_inputs, training_classes, testing_classes) = train_test_split(
    feature_values_scaled, classe_values, train_size=0.75, random_state=1
)
```

```python
classsifier = DecisionTreeClassifier(random_state=1)
classsifier.fit(training_inputs, training_classes)
```

```python
from pydotplus import graph_from_dot_data
from IPython.display import Image
```

```python
# Visualize the decision tree
dot_data = StringIO()
tree.export_graphviz(classsifier, out_file=dot_data, feature_names=feature_names)
graph = graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
classsifier.score(testing_inputs, testing_classes)
```

### improve with Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classsifier = RandomForestClassifier(n_estimators=10, random_state=1)
cvs = cross_val_score(classsifier, feature_values_scaled, classe_values, cv=10)
```

```python
cvs.mean()
```

# KNN

```python
from sklearn import neighbors
```

```python
classsifier = neighbors.KNeighborsClassifier(n_neighbors=10)
cvs = cross_val_score(classsifier, feature_values_scaled, classe_values, cv=10)
```

```python
cvs.mean()
```

# Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
```

```python
scaler = preprocessing.MinMaxScaler()
feature_values_minmax = scaler.fit_transform(feature_values_scaled)

classsifier = MultinomialNB()
cvs = cross_val_score(classsifier, feature_values_minmax, classe_values, cv=10)
```

```python
cvs.mean()
```

# Logistic Regression

```python
classsifier = LogisticRegression()
cvs = cross_val_score(classsifier, feature_values_scaled, classe_values, cv=10)
```

```python
cvs.mean()
```

# Neural Networks

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from scikeras.wrappers import KerasClassifier
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(6, input_dim=4, kernel_initializer="normal", activation="relu"))
    model.add(Dense(1, kernel_initializer="normal", activation="sigmoid"))
    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model
```

```python
estimator = KerasClassifier(model=create_model, epochs=100, verbose=0)
cvs = cross_val_score(estimator, feature_values_scaled, classe_values, cv=10)
```

```python
cvs.mean()
```
